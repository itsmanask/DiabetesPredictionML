```python
# 1. Calculate Mean, Median, and Mode
mean_values = df.mean(numeric_only=True)
median_values= df.median(numeric_only=True)
mode_values= df.mode(numeric_only=True).iloc[0]
```

```python
#2.Adding new row
newrow = {
    'Pregnancies': 2,
    'Glucose': 120,
    'BloodPressure': 70,
    'SkinThickness': 20,
    'Insulin': 80,
    'BMI': 25.3,
    'DiabetesPedigreeFunction': 0.5,
    'Age': 29,
    'Outcome': 0
}
df = pd.concat([df, pd.DataFrame([newrow])], ignore_index=True)
print(newrow)

```

```python
#3. Adding new column
df['RiskLevel'] = df['Glucose'].apply(lambda x: 'High' if x > 125 else 'Low')
```

```python
# 4. Remove a row 
df = df.drop(index=0).reset_index(drop=True)
```

```python
# 5. Remove a column 
df = df.drop(columns=['SkinThickness'])
```

```python
# 6. Fill zero values in 'Insulin' and 'BMI' with their respective mean
df['Insulin'] = df['Insulin'].replace(0, df['Insulin'].mean())
df['BMI'] = df['BMI'].replace(0, df['BMI'].mean())
```

```python
df.head(), mean_values, median_values, mode_values
```

```python
# Rename specific columns
df = df.rename(columns={'Age': 'Patient_Age', 'BMI': 'Body_Mass_Index'})
```

```python
# Sort by a single column
df = df.sort_values(by='Patient_Age')

# Sort by multiple columns
df = df.sort_values(by=['Patient_Age', 'Body_Mass_Index'], ascending=[True, False])
```

```python
# Filter Rows (Conditional Selection)
# Select all rows where BMI > 30
high_bmi = df[df['Body_Mass_Index'] > 30]

# Patients older than 50 with diabetes
older_diabetic = df[(df['Patient_Age'] > 50) & (df['Outcome'] == 1)]
```

```python
# Create a new column 'Risk' based on BMI
df['Risk'] = ['High' if x > 30 else 'Normal' for x in df['Body_Mass_Index']]
```

```python
# Reset the index after deleting rows
df = df.reset_index(drop=True)
```

```python
# Find and remove duplicate rows
duplicates = df[df.duplicated()]
df = df.drop_duplicates()
```

```python
# --- Finally Print the Updated Data ---
print(df)
```
